```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("./Crop_recommendation.csv")
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
modified_dataset=dataset
```

```python
modified_dataset.label=le.fit_transform(modified_dataset.label)
modified_dataset
```

```python
modified_dataset=modified_dataset.sample(frac=1).reset_index(drop=True)
modified_dataset.to_csv("Modfied_Dataset.csv",index=False)
modified_dataset
```

```python
np.random.seed(42)

data = modified_dataset.drop('label',axis=1)
target = modified_dataset['label']
```
